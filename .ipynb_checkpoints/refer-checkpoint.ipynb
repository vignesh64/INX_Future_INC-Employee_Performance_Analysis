```python
import pandas as pd
import numpy as np
```

```python
filename = "data/RAW/INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls"
```

```python
df = pd.read_excel(filename)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
enc = LabelEncoder()
orc = OrdinalEncoder()
for i in (2,3,4,5,6,7,16,26):
    df.iloc[:,i] = enc.fit_transform(df.iloc[:,i])
df.head()
```

```python
df.dtypes
```

```python
del df['EmpNumber']
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
```

```python
x  = df.iloc[:,:-1]
y = df.iloc[:,-1:]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,stratify=y,test_size = 0.25,random_state = 41)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
sc_1 = StandardScaler()
```

```python
scaled_x = sc_1.fit_transform(x_train)
scaled_x_test = sc_1.transform(x_test)
```

```python
sm = SMOTE()
```

```python
x_train_sm,y_train_sm = sm.fit_resample(scaled_x,y_train)
```

```python
y_train_sm.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
model_lr = LogisticRegression()
```

```python
model_lr.fit(x_train_sm,y_train_sm)
```

```python
lr_pred = model_lr.predict(scaled_x_test)
```

```python
accuracy_score(y_test,lr_pred)
```

```python
print(classification_report(y_test,lr_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train_sm,y_train_sm)
```

```python
rf_pred = rf.predict(x_test)
```

```python
accuracy_score(y_test,rf_pred)
```

```python
print(classification_report(y_test,rf_pred))
```

```python
from xgboost import XGBClassifier
from catboost import CatBoostClassifier 
```

```python
xgb_clf = XGBClassifier()
```

```python
xgb_clf.fit(x_train_sm,y_train_sm)
```

```python
xg_pred = xgb_clf.predict(scaled_x_test)
```

```python
accuracy_score(y_test,xg_pred)
```

```python
print(classification_report(y_test,xg_pred))
```

```python
cat_clf = CatBoostClassifier()
```

```python
cat_clf.fit(x_train_sm,y_train_sm)
```

```python
cat_pred = cat_clf.predict(scaled_x_test)
```

```python
accuracy_score(y_test,cat_pred)
```

```python
print(classification_report(y_test,cat_pred))
```
