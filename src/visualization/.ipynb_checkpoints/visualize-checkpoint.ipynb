## Department wise performance

```python
#considering the top 3 factors to visualize the performances in each department
def plots(data):
    plt.figure(figsize=(20,5))
    plt.subplot(1,3,1)
    sns.countplot(data.EmpEnvironmentSatisfaction)
    #plt.legend(loc = 'best')
    plt.subplot(1,3,2)
    sns.countplot(data.YearsSinceLastPromotion, hue = data.PerformanceRating)
    plt.legend(loc = 'best')
    plt.subplot(1,3,3)     
    sns.countplot(data.EmpLastSalaryHikePercent, hue = data.PerformanceRating)
    plt.legend(loc = 'best')
```

#### Sales

```python
Sales = data[data.EmpDepartment == 'Sales']
plots(Sales)
```

1. The employee enviromnment satisfaction count are more in top two ratings but quite noticeable people are less satisfied with the environment.
2. As the years since the last promotion increases the performnace decreases, 0 - 1 yrs since promotion performances are high.
3. The green bar which represent the top performance is high when the hike is above 20% and the performances is just average for 11 - 16% hike percent.

#### Development

```python
Development = data[data.EmpDepartment == 'Development']
plots(Development)
```

1. The Employee environment satisfaction rating 3 is more in this department, around 70 employees are less satisfied with the environment.
2. 0-2 yrs since the promotion are performing well. As the years increases the performances are questionable.
3. Performnaces of the employees whose hike percent is above 20 are working well whereas below that are mostly average performances.

#### Research & Development

```python
R_D = data[data.EmpDepartment == 'Research & Development']
plots(R_D)
```

1. There are more employees in this department who are quite satisfied with the company environment.
2. Same as the other departments the 0 -2 yrs since the promotion are performing well.
3. The performance of the employee who has hike abov 20% are performing really well than of lesser percentage.

#### Human Resources

```python
HR = data[data.EmpDepartment == 'Human Resources']
plots(HR)
```

1. Employees in this department are moderately satisfied with the evnironment.
2. There are few poor performances when it's been only a  year or two since the Promotions in this department.
3. Employees with less hike percent are performing moderately in this department.

#### Finance

```python
Finance = data[data.EmpDepartment == 'Finance']
plots(Finance)
```

1. There more number of employees who are satisfied with the environment.
2. Employees who have been promoted a year ago are performing low, probably they must be panalized. They are few high performancea for increase in years since the promotion
3. Employee with hike percent above 20 are performing well, lesser than that are either moderate or poor performnaces.

#### Data science

```python
Data_science = data[data.EmpDepartment == 'Data Science']
plots(Data_science)
```

1. Most of the employees are satisfied with the environment.
2. Performnaces are not much damaged even if the years since last promotion are higher, performances is either 3 or 4.
3. Performances are good irrespective of the salary hike percentage.
