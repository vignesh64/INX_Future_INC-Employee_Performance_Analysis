## Data Processing

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

### Reading Raw data

```python
data_path = 'D:/iabac/data/raw.xls'
df = pd.read_excel(data_path)
```

```python
df.columns
```

### Label Encoding

Since there are categorical features(Non- Numerical) in the dataset these columns to Nuerical columns beacuse the model only to be trained on numerical values.

```python
enc = LabelEncoder()
for i in (2,3,4,5,6,7,16,26): #features that are to be encoded 
    df.iloc[:,i] = enc.fit_transform(df.iloc[:,i])
```

```python
df
```

The EmpNumber would be of no use in the prediction and training of model so deleting the column.

```python
del df['EmpNumber']
```

```python
#creating the processed data to model and predict
df.to_excel('data/processed.xls',index = False)
```
